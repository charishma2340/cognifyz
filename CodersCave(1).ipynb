```python
# Importing Libraries
import numpy as np
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.offline import iplot
import seaborn as sns
```

```python
# Importing dataset
df = pd.read_csv('/content/dataset(1).csv', encoding='cp1252', error_bad_lines=False)
```

```python
#printing dataset
df
```

```python
# preparing data
pd.set_option('display.max_columns', None)
df = df.rename(columns={'iday':'day','imonth':'month','iyear':'year','country_txt':'country_name','nkill':'killed',
                    'nwound':'wounded','attacktype1_txt':'attack type','weaptype1_txt': 'weapon type'})
df[['month','day']] = df[['month','day']].replace(0,1)#some incidents have value 0 for month/day so we replace with 1
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])
df.sample(5)
```

```python
# statistics measures of each column
df.describe()
```

```python
#calculating missing data in each column
df[['country_name','year','country','city','killed','wounded']].isnull().sum()
```

```python
# Calculating terrorist attacks per country
country_attacks = df['country_name'].value_counts().reset_index()
country_attacks = country_attacks.rename(columns={'index': 'country_name', 'country_name': 'incidents'})

head_countries = country_attacks.head(20)[::-1]

# Plotting
fig = go.Figure(go.Bar(
            y=head_countries['country_name'],
            x=head_countries['incidents'],
            orientation='h'))

fig.update_layout(title=dict(text='Total terrorist incidents per Country 1970-2017', font=dict(size=25)),
                  xaxis_title="Total incidents",
                  yaxis_title="Country")

fig.show()
```

```python
# Calculating terrorist attacks per city
city_attacks = df['city'].value_counts().reset_index().iloc[1:21,:]
city_attacks.rename(columns={'index': 'city', 'city': 'incidents'}, inplace=True)
head_city = city_attacks.head(20)[::-1]

# Plotting
fig = go.Figure(go.Bar(
            y=head_city['city'],
            x=head_city['incidents'],
            orientation='h'))

fig.update_layout(title=dict(text='Total terrorist incidents per City 1970-2017', font=dict(size=25)),
                  xaxis_title="Total incidents",
                  yaxis_title="City")

fig.show()
```

```python
#reading the online data
population_data = pd.read_html('https://www.worldometers.info/world-population/population-by-country/')[0]
population_data = population_data.rename(columns={'Country (or dependency)':'country_name',
                                                  'Population  (2023)':'population'})
country_attacks = country_attacks.merge(population_data[['country_name','population']], on='country_name')
country_attacks['adj_incidents'] = (country_attacks.incidents / country_attacks.population) * 100000
#finding the countries with the highest range of incidents
head_adj_countries = country_attacks.sort_values(by='adj_incidents', ascending=False).head(20)[::-1]

fig = go.Figure(go.Bar(
            y=head_adj_countries.country_name,
            x=head_adj_countries.adj_incidents,
            orientation='h'))

fig.update_layout(title=dict(text='Countries with most incidents per 100,000 inhabitants 1970-2017',
                             font=dict(size=25)),
                xaxis_title="Incidents per 100,000 inhabitants",
                yaxis_title="Country",)

fig.show()
```

```python
#time series analysis
frequencies_year = df[['iyear','eventid']].groupby('iyear').count().reset_index()
frequencies_year = frequencies_year.rename(columns={'eventid':'number of incidents'})

fig = px.line(frequencies_year, x="iyear", y="number of incidents")
fig.update_layout(title=dict(text='Terrorist incidents over time 1970-2017', font=dict(size=25)),
                xaxis_title="Year",
                yaxis_title="Number of incidents",)

fig.show()
```

```python
#check the average sum amount of incidents per day of the week
df['day_week'] = df.date.dt.day_name()
week_day_incidents = df[['day_week','eventid']].groupby('day_week').count().reset_index()\
                                .sort_values(by='eventid', ascending=False)
```

```python
#Visualizing data
t1 = go.Bar(
    x = week_day_incidents.day_week,
    y = week_day_incidents.eventid,
    name = 'Incidents per week day',
)

layout = dict(height=400,
              width=1000,
              title = 'Total incidents per week day',
              yaxis = dict(title = 'Number of incidents'),
              xaxis= dict(title='Day of week'),
              )

fig = dict(data = t1, layout=layout)

iplot(fig)
```

```python
#calculating terrorist incidents per region
frequencies_year_region = df[['iyear','eventid','region_txt']].groupby(['iyear','region_txt']).count().reset_index()
frequencies_year_region = frequencies_year_region.rename(columns={'eventid':'number of incidents',
                                                                 'region_txt':'Region Name'})

fig = px.line(frequencies_year_region, x="iyear", y="number of incidents", color='Region Name')
fig.update_layout(title=dict(text='Terrorist incidents per region over time 1970-2017', font=dict(size=25)),
                xaxis_title="Year",
                yaxis_title="Number of incidents",)

fig.show()
```

```python
#finding the total number of killed and wounded for every year
casualties_year = df[['iyear', 'nkill', 'nwound']].groupby('iyear').sum().reset_index()
casualties_year = pd.melt(casualties_year,
                            id_vars='iyear',
                            value_vars=['nwound', 'nkill'],
                            var_name='casualty type',
                            value_name='amount')

fig = px.line(casualties_year, x="iyear", y="amount", color='casualty type', color_discrete_sequence=['orange','deeppink'])
fig.update_layout(title=dict(text='Terrorist incidents casualties over time 1970-2017', font=dict(size=25)),
                xaxis_title="Year",
                yaxis_title="Number of casualties")

fig.show()
```

```python
# Calculating terrorist incidents killings
killings_year_region = df[['iyear','nkill','region_txt']].groupby(['iyear','region_txt']).sum().reset_index()
killings_year_region = killings_year_region.rename(columns={'region_txt':'Region Name'})

fig = px.line(killings_year_region, x="iyear", y="nkill", color='Region Name')
fig.update_layout(title=dict(text='Terrorist incidents killings per region over time', font=dict(size=25)),
                xaxis_title="Year",
                yaxis_title="Number of killings",)

fig.show()
```

```python
# Calculating terrorist incidents wounded
wounded_year_region = df[['iyear','nwound','region_txt']].groupby(['iyear','region_txt']).sum().reset_index()
wounded_year_region = wounded_year_region.rename(columns={'region_txt':'Region Name'})

fig = px.line(wounded_year_region, x="iyear", y="nwound", color='Region Name')
fig.update_layout(title=dict(text='Terrorist incidents wounded per region over time', font=dict(size=25)),
                xaxis_title="Year",
                yaxis_title="Number of wounded",)

fig.show()
```
